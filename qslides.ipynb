## 주요변경점

**2학년 전공 교과목**

- 수강 신청 경쟁이 치열할 것으로 예상되는 과목위주로 분반 개설.
- 이번 학기에는 「회귀분석」과 「데이터관리 및 처리」 과목이 각각 2개 분반으로 운영. (이 중 하나의 분반은 통계학과 2학년 학생을 대상으로 이미 일괄 수강신청이 완료되었음.) 

## 주요변경점

**3-4학년 전공 교과목**

- 개설 과목 수를 확대하여 다양한 선택지를 제공. 
- 3학년과 4학년 과목 간 시간표 중복을 최소화하여 수강 편의성을 높임. 
- 학생들의 선호도가 높은 과목을 중심으로 개설.

## FAQ

통계학과 2학년 학생입니다. 이번 학기에는 **「데이터관리 및 처리」**를 수강할 계획이 없고, 내년에 3학년이 된 후에 듣고자 합니다. 저처럼 이번 학기에 이 과목을 듣지 않아도 되나요?

> 답변: 네, 수강신청 기간에 해당 과목을 제외하고 다른 과목으로 신청하시면 됩니다. 다만, 3학년 시점에는 「데이터관리 및 처리」의 수강이 보장되지 않을 수 있으니 이 점 유의하시기 바랍니다.

## FAQ

통계학과 2학년 학생입니다. **「데이터관리 및 처리」**를 수강하고 싶지만, 개인적인 사정으로 인해 일괄 신청된 2분반이 아닌 1분반 수업을 듣고 싶습니다. 가능한가요?

> 답변: 네, 수강신청 기간에 2분반을 제외하고 1분반으로 직접 신청하시면 됩니다. 다만, 1분반의 경우 통계학과 2학년 학생에게 수강이 보장되지 않으므로 신청 결과에 따라 수강이 불가할 수도 있습니다. 이 점 유의해 주세요.

## FAQ

비통계학과 2학년 학생입니다. 주요 전공 교과목에 대해 비통계학과 학생의 수강권이 통계학과 학생에 비해 제한되는 것 같습니다. 향후 개선 계획이 있을까요?

> 답변: 비통계학과 학생들의 수강 기회를 확대하기 위한 방안을 내부적으로도 고민하고 있습니다. 다만 현재 본부 시스템상 본과 학생 외에는 일괄 수강신청 기능 자체가 제공되지 않아, 제도적으로 적용이 어려운 상황입니다. 향후 본부 시스템이 개선된다면, 통계학과에서도 이에 맞춰 제도를 개선할 의지가 있습니다. 현 시점에서는 개별적으로 1분반에 수강신청해 주시기 바랍니다.

## FAQ

통계학과 3학년 학생입니다. 개인적인 사정으로 인해 「데이터관리 및 처리」 2분반을 수강하고 싶습니다. 가능할까요?

> 답변: 2분반은 통계학과 2학년 학생을 위한 전용 분반으로 운영되고 있습니다. 따라서 3학년 학생께서는 1분반으로 수강신청해 주시기 바랍니다.

## FAQ

통계학과 3학년입니다. 지금까지 통계학입문, 선형대수학, 수리통계학1·2, 회귀분석1, 기계학습을 수강했는데, 전공필수 6학점을 채우려면 어떤 과목을 더 들어야 하나요?

> 답변: 기존에는 「데이터관리 및 처리」, 「데이터과학 입문」, 「확률의 개념 및 응용」, 「회귀분석」 중 2과목을 이수해야 했습니다. 하지만 최근 교수회의에서 「통계학입문」을 전공필수로 인정하기로 결정되어, 해당 과목을 이수한 경우에는 위 과목 중 1과목만 추가로 수강하시면 됩니다.

## 복수전공,부전공,연계전공 소개

**복수전공·부전공·연계전공의 졸업상 이점**

- 복수전공, 부전공, 연계전공을 통해 융합인재로의 성장, 진로 확장의 기회, 다학제적 사고력 향상 등의 발전 가능성을 기대할 수 있음. 
- 복수전공, 부전공, 연계전공 이수 시 겹치는 교과목의 중복 인정이 가능하며, 이를 활용하면 전체 졸업학점을 보다 효율적으로 충족할 수 있음.
